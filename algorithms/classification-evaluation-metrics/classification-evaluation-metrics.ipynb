```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../datasets/ckd.csv")
df.tail()
```

```python
df = df.iloc[:,2:]
```

```python
categorical_cols = ['rbc','pc','pcc','ba','htn','dm','cad','appet','pe','ane','classification']
for i in categorical_cols:
    df[i] = df[i].astype('category')
    un = df[i].unique()
    df[i] = np.where(df[i] == un[0],1,0)
df
```

```python
X = df.iloc[:,:24]
y= df.iloc[:,24]
```

##### Select a training model
Model chosen: Logistic Regression

##### Divide into train-test-split

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, solver='lbfgs').fit(X_train, y_train)
```

##### Classification accuracy

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
# print the first 25 true and predicted responses
print('True:', y_test.values[0:25])
print('False:', y_pred[0:25])
```

##### Confusion matrix

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

TP: 35  TN: 12  FP: 1  FN: 0

##### Confusion matrix metrics

```python
TP = 35
TN = 12
FP = 1
FN = 0
print((TP + TN) / float(TP + TN + FP + FN))
print(accuracy_score(y_test, y_pred))
```

Classification error:

```python
classification_error = (FP + FN) / float(TP + TN + FP + FN)

print(classification_error)
print(1 - accuracy_score(y_test, y_pred))
```

Sensitivity:

```python
from sklearn.metrics import recall_score
sensitivity = TP / float(FN + TP)
print(sensitivity)
print(recall_score(y_test, y_pred))
```

Specificity:

```python
specificity = TN / (TN + FP)

print(specificity)
```

Precision:

```python
from sklearn.metrics import precision_score
precision = TP / float(TP + FP)

print(precision)
print(precision_score(y_test, y_pred))
```

##### ROC Curves

```python
y_pred_prob = clf.predict_proba(X_test)[:, 1]
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_prob)

plt.plot(fpr, tpr)
plt.rcParams['font.size'] = 12
plt.title('ROC curve')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
```

##### AUC

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test, y_pred_prob))
```

Cross validated AUC:

```python
from sklearn.cross_validation import cross_val_score
cross_val_score(clf, X_test, y_test, cv=10, scoring='roc_auc').mean()
```
