```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
ckd = pd.read_csv("/home/ijaju/Documents/datasets/ckd.csv")
ckd = ckd.iloc[:,2:]
```

```python
ckd.columns
```

```python
ckd=ckd.iloc[:,[0,1,2,3,4,9,10,11,12,13,14,15,16,17,24]]
ckd['classification'] = np.where(ckd['classification']=='ckd',0,1)
```

```python
X = ckd.iloc[:,0:13].values
y = ckd.iloc[:,14].values

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn import svm
clf = svm.SVC(gamma='auto')
clf.fit(X_train, y_train)  
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
clf = svm.SVC(C=0.0000001,gamma='auto')
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
clf = svm.SVC(C=1000000,gamma='auto')
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
C = [0.0000001,1000000]
for c in C:
    clf_ = svm.SVC(C=c,gamma='auto')
    clf_.fit(X_train, y_train)
    y_pred=clf_.predict(X_test)
    print("For C =",c,", accuracy =",accuracy_score(y_test,y_pred))
```

```python
clf = svm.SVC(kernel='linear',gamma='auto')
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
clf = svm.SVC(kernel='poly',gamma='auto')
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
clf = svm.SVC(kernel='sigmoid',gamma='auto')
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
kernel = ['linear','poly','sigmoid']
for k in kernel:
    clf = svm.SVC(kernel=k,gamma='auto')
    clf.fit(X_train, y_train)
    y_pred=clf.predict(X_test)
    print("For kernel =",k,", accuracy =",accuracy_score(y_test,y_pred))
```

```python
clf = svm.SVC(gamma='scale')
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
clf = svm.SVC(gamma=3000)
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
gamma = [0.001,0.1,1,100,3000]
for g in gamma:
    clf = svm.SVC(gamma=g)
    clf.fit(X_train, y_train)
    y_pred=clf.predict(X_test)
    print("For gamma =",g,", accuracy =",accuracy_score(y_test,y_pred))
```

### SVM for Regression
Using automobiles dataset

```python
headers = ['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels',
           'engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders',
           'engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg',
           'highway-mpg','price']
auto = pd.read_csv("/home/ijaju/Documents/datasets/automobiles.data",header=None,names=headers)
```

```python
print(auto.shape)
auto=auto.replace('?', np.nan)
auto = auto.dropna(axis=0, how="any")
print(auto.shape)
```

```python
auto.tail()
```

```python
str_cols = ['make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type',
            'num-of-cylinders','fuel-system']
for i in auto.columns:
    if i in str_cols:
        e = auto[i].unique()
        ind=0
        for j in e:
            auto[i].loc[auto[i]==j] = ind
            ind+=1
```

```python
auto
```

```python
X = auto.iloc[:,:24]
y=auto.iloc[:,25]

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.svm import SVR
clf = SVR(gamma='auto', C=1.0, epsilon=0.2)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
clf.predict(X_test)
clf.score(X_test,y_test)
```
